# 1. Write a function to check whether a string is a palindrome.

```python
def palindrome_detector(word):
    word = word.replace(" ", "")
    pal = list(word)
    pal.reverse()
    reverse = "".join(pal)
    print("Reverse Word:",reverse)
    if word == reverse:
        print("Palindrome detected")
    else:
        print("Not a Palindrome")
    return

word = input("Enter a word: ")
palindrome_detector(word)
```

# 2. Write a function that checks whether a number is prime or not.

```python
def is_prime(n):
    if n > 1:
        for i in range(2,n):
            if (n % i) == 0:
                print(n, "is not a prime number")
                print(i, "times", n//i, "is", n)
                break
        else:
            print(n, "is a prime number")
    else:
        print(n, "is not a prime number")
                
                
number = int(input("Enter a number: "))
is_prime(number)
```

# 5 Load Rick Astley - Never Gonna Give You Up Lyrics _ AZLyrics.com.html

# Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import requests
import string
from bs4 import BeautifulSoup


page = requests.get("https://www.azlyrics.com/lyrics/rickastley/nevergonnagiveyouup.html")
soup = BeautifulSoup(page.content, 'html.parser')
output = soup.find_all('div')[19].get_text()
output.strip()

clean_2 = output.translate(str.maketrans('', '', string.punctuation))
clean_2 = clean_2.lower()

word_list_2 = clean_2.split()
unique_words_2 = set(word_list_2)
print("Total Unique words:", len(unique_words_2))
```

```python
print(clean_2)
```

```python
import io
buf = io.StringIO(clean_2)
with buf as file:
    for line in buf:     
        if line.startswith('never'):
            print(line)
```

```python
b = {}

for word in unique_words_2:
    b[word] = 0
    
for word in word_list_2:
    b[word] = b[word] + 1
b
```

```python
import io
buf = io.StringIO(clean_2)
with buf as file:
    data = file.read()
    
d = Counter(data.split()).items()
df = pd.DataFrame(d, columns=['word', 'count'])
df
```

```python
dfsorted = df.sort_values(by=['count'], ascending=False)
dfsorted = dfsorted.reset_index(drop=True)
dfsorted
```

```python
wordcount = dfsorted

%matplotlib inline
plt.figure(figsize=(13,22))
sns.set(style="darkgrid")

ax = sns.barplot(y='word', x='count', data=wordcount.reset_index(), color="lightcoral")
ax.set(xlabel='Word Frequency', ylabel='Unique Words')
labels = dfsorted['count']
for i, v in enumerate(labels):
    ax.text(v + 0.5, i + .26, str(v), color='black', fontweight='roman')
ax.locator_params(axis='count', integer=True)
plt.title("Word Count Histogram")
plt.show()
```
